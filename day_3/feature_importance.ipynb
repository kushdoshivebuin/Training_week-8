```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
```

```python
X, y = make_classification(n_samples=5, n_classes=2, n_features=2, n_informative=2, n_redundant=0, random_state=0)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X, y)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf)
```

```python
clf.feature_importances_
```

Formula for calculation of Feature Importance :

n<sub>i</sub> = (N_t/N) * [impurity - ((N_t_r/N_t) * right_impurity) - ((N_t_l/N_t) * left_impurity)]

where   N_t = number of samples in that node

        N = total number of rows in the data
        impurity = gini impurity of that node
        N_t_r = right child node's number of samples
        right_impurity = gini impurity of right child node
        N_t_l = left child node's number of samples
        left_impurity = gini impurity of left child node

Calculating for X[1] :

n<sub>1</sub> = (5/5) * [0.48 - ((4/5) * 0.375) - ((1/5) * 0)]

n<sub>1</sub> = 1 * (0.48 - 0.3 - 0)

n<sub>1</sub> = 0.18

Similarly for X[0] :

n<sub>0</sub> = 0.8 * (0.375 - 0 - 0)

n<sub>0</sub> = 0.3

0<sup>th</sup> = n<sub>0</sub> / (n<sub>0</sub> / n<sub>1</sub>)

0<sup>th</sup> = 0.3 / (0.3 + 0.18)

0<sup>th</sup> = 0.625

Similarly for 1<sup>st</sup> = 0.375
